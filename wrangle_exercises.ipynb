```python
# Calculate q1, q3 and iqr for income

q1 = df.annual_income.quantile(0.25)
q3 = df.annual_income.quantile(0.75)
iqr = q3 - q1
k = 1.5

# calculate upper and lower ranges
upper_bound_income =  q3 + k * iqr
lower_bound_income =  q1 - k * iqr

upper_bound_income
```

```python
df[df.annual_income > upper_bound_income]
```

```python
bounds_dict = {}

for col in df.dtypes[df.dtypes != 'object'].index:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    upper_bound =  q3 + k * iqr
    lower_bound =  q1 - k * iqr
    
    bounds_dict[col]['upper_bound'] = upper_bound
    bounds_dict[col]['lower_bound'] = lower_bound

df = df[(df[col] < upper_bound) & (df[col] > lower_bound)]
```
